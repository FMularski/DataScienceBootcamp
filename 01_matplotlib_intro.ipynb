```
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
```

```
fig = plt.figure()
fig.show()
```

```
plt.plot([0, 1], [0, 3])
```

```
plt.plot([0, 1], [0, 3], label='line', color='green')
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('Wykres')
plt.legend()
```

```
plt.plot([0, 1, 2], [0, 3, 2], label='line', color='green')
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('Wykres')
plt.legend()
```

```
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)
plt.plot(x, x_sin)
plt.plot(x, x_cos)
```

```
plt.subplot(121)
plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(122)
plt.plot([0, 1, 2], [3, 1, 3], 'r--')
```

```
fig = plt.figure(figsize=(12, 8))
plt.subplot(131)
plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(132)
plt.plot([0, 1, 2], [3, 1, 3], 'r--')
plt.subplot(133)
plt.plot([0, 1, 2], [3, 1, 3], 'go')

# fig.show()
```

```
plt.bar(x=['a', 'b', 'c', 'd'], height=[1.2, 3.2, 1.2, 5.2])
```

```
plt.barh(y=['a', 'b', 'c', 'd'], width=[1.2, 3.2, 1.2, 5.2])
```

```
x1 = np.random.randn(300)
x2 = np.random.randn(300)

plt.scatter(x1, x2)
```

```
x3 = np.random.randn(300) * 50
plt.scatter(x=x1, y=x2, s=x3, alpha=0.5)
```
